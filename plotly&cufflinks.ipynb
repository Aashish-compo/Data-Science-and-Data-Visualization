```python
# Creating DATA 
df=pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
```

```python
df.head()
```

```python
df2= pd.DataFrame({'Category':['A', 'B', 'C'], 'Values': [32,44,53]})
```

```python
df2
```

```python
df.iplot()
```

```python
#Scatter Plot
df.iplot(kind='scatter', x='A',y='B', mode='markers',size=15)
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

## 3D Surface Plot

```python
df3=pd.DataFrame({'x':[1,2,3,4,5], 'y':[10,20,30,20,10], 'z':[5,4,3,2,1]})
```

```python
# Surface type ko graph... colorscale aafaile set gareko...yesko lagi documentation hernu parxa, kinaki yesko colorscale matplotlib sanga mildaina...yesko aafnai xa...SO DOCUMENTAION IS THE BEST
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
# Histogram
df['B'].iplot(kind='hist', bins=15)
```

```python
#Spread type ko data visualization 
df[['A', 'B']].iplot(kind='spread')
```

```python
# Bubble plot
## similar to scatter plot but the size will be accroding to the other variable..In the below case size of bulle will change according the value of 'C'
df.iplot(kind='bubble', x='A', y='B', size='C')
```

```python
# Scatter_matrix.... NOTE: similar to seaborn pairplot
df.scatter_matrix()
```
