Everything about 2D array

```python
arr_2d = np.array([[1,2,3], [6,9,5], [7,4,0]])
```

```python
arr_2d
```

Grabbing an element from the 2D array:

```python
arr_2d[0][2]
```

```python
arr_2d[1,2]
```

Grabbing the sub-matrix from already existing matrix:

Reading arr_2d[:2,1:] as grabbing everything up to row 2 and grab from column 1 onwards.

```python
arr_2d[:2,1:]
```

```python
arr_2d[1:,:]
```

Conditional Selection

```python
new_arr = np.arange(1,11)
```

```python
new_arr
```

```python
bool_arr = new_arr > 5
```

```python
bool_arr 
```

```python
 new_arr[bool_arr]
```

But we can do all this conditional selection in one line of code.

```python
new_arr[new_arr>5]
```

Final practice of slicing the 2D array:

```python
ar_2d = np.arange(50).reshape(5,10)
```

```python
ar_2d
```

Below shows the slicing of different sub-matrix of above Matrix.

```python
ar_2d[1:3,3:5]
```

```python
ar_2d[2:,5:]
```
