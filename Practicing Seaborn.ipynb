# Seaborn
* Seaborn is a statistical plotting library, 
* It has beautiful default styles.
* It also is designed to work very well with pandas dataframe objects.

### 1. Distribution Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde= False, bins=40)
```

```python
sns.jointplot(x='total_bill' ,y='tip' ,data=tips, kind='kde')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
  sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

### 2. Categorical Plots

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True,hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

### 3. Matrix Plot

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tc = tips.corr()
```

```python
# For showing data in heatmap, data should be in matrix form.
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fp= flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor='white', lw=2.5)
```

```python
sns.clustermap(fp,standard_scale=1)
#For normalizing, we can use standard_scale arguement
```

### 4. Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

### 5. Regression Plot

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```
