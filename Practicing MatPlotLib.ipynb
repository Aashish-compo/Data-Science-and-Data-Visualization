# Matplotlib

```python
  import matplotlib.pyplot as plt
```

```python
 #This code helps us to see the plot inside jupyter notebook.
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

There are two method to show the plot. ONE is using functional approach and the SECOND is using object oriented approach.

```python
#Functional method
plt.plot(x,y)
# plt.show()   If not in jupyter notebook, this extra line of code is also needed.
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #(left, bottom, width, height)
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y**2,x)
axes2.set_title('SMALLER PLOT')
```

For control over the number of rows and columns using object oriented appraoch, this method is helpful.
THIS OO APPRAOCH IS FAST AS WE DON'T HAVE TO ADD AXES OF LEFT,BOTTOM,WIDTH,HEIGHT.

```python
fig,axes = plt.subplots(nrows=3, ncols=3)

#axes.plot(x,y)
plt.tight_layout() #For fixing the issures of plots overlapping.
```

```python
fig,axes = plt.subplots(nrows =1, ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
axes # This axes is just the array of matplotlib axes.
```

Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2)) # Here figsize is the tuple and it is the width and height of figure in inches. 

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label = 'X,Y')
ax.plot(x**2,y, label = 'X^2,Y')

ax.legend(loc = 0)
```

```python
 # We can also do the exact same thing using subplot as well.
    
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(4,4))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
# Saving figure
fig.savefig('my_pictuer.png', dpi= 200)
```

PLOT APPEARANCE

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'red', linewidth = 3, alpha = 0.5, linestyle= '-.', marker = 'o', markersize= 20, markerfacecolor= 'blue', markeredgewidth= 4,markeredgecolor= 'green')

# linewidth/lw for controlling the width and alpha for controlling the tranparency of line and linestyle/ls for styling the line.
```

Control over axis appearance

```python
# plot range
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'red', lw = 3, alpha = 0.5, ls= '--')

ax.set_xlim(0,1)
ax.set_ylim(0,2)
```
